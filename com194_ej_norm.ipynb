```python
import warnings
import math
import numpy as np # to use numpy arrays instead of lists
import pandas as pd # DataFrame (table)
import matplotlib.pyplot as plt # to plot
```

```python
x = np.arange(110)
y = 3*np.sin(x)-2*np.cos(x)+4
yMin=np.min(y)
print("Min value: ", yMin)

y_scalar=(y-np.min(y))/(np.max(y)-np.min(y))
print("y_scalar var: ", np.var(y_scalar))
y_scalar2=y/np.max(y)
print("y_scalar2 var: ", np.var(y_scalar2))
y_exp=x**(1/2)
y_exp_scale=(y_exp-np.min(y_exp))/(np.max(y_exp)-np.min(y_exp))
y_exp_scale2=(y_exp)/(np.max(y_exp))

plt.figure()
#plt.plot(y)
#plt.plot(y_scalar)
#plt.plot(y_scalar2)
#plt.plot(y_exp)
plt.plot(y_exp_scale+1)
plt.plot(y_exp_scale2)
plt.grid(color='b',linestyle='dashed')
plt.title('SINE', fontweight='bold')
plt.xlabel('Tiempo')
plt.ylabel('Valor')
plt.legend(loc='upper left')
plt.show()
```
